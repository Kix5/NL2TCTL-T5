```python
import csv
from typing import Tuple, Dict, Sequence
import re
import math
import random

def load_dataset(path: str):
    data = []
    with open(path, newline='', encoding='utf-8') as f:
        reader = csv.DictReader(f)  # uses the first row as fieldnames
        for row in reader:
            nl  = row.get('refined NL') or row.get('nl')
            tctl = row.get('TCTL') or row.get('tctl')
            if nl and tctl:
                data.append((nl.strip(), tctl.strip()))
    return data

AP_PATTERN = re.compile(r"[a-zA-Z_][a-zA-Z0-9_]*")

# Replace atomic propositions with placeholders
def lift_ap(formula: str) -> Tuple[str, Dict[str, str]]:
    operators = {"A", "E", "AF", "AG", "EF", "EG", "U", "¬", "!", "->", "→",
                 "∧", "and", "∨", "or", "<=", ">=", "<", ">", "[", "]", "(", ")", "."}
    mapping: Dict[str, str] = {}
    reversed_mapping: Dict[str, str] = {}
    def repl(match: re.Match[str]) -> str:
        token = match.group(0)
        if token.isdigit() or token in operators:
            return token
        if token in reversed_mapping:
            return reversed_mapping[token]
        idx = len(mapping) + 1
        placeholder = f"prop_{idx}"
        mapping[placeholder] = token
        reversed_mapping[token] = placeholder
        return placeholder
    lifted = AP_PATTERN.sub(repl, formula)
    return lifted, mapping

# Normalise spacing and operator case in a formula
def normalize_formula(formula: str) -> str:
    expr = formula.strip()
    expr = re.sub(r"([()[\]])", r" \1 ", expr)
    expr = re.sub(r"(<=|>=|<|>|=|\\.|!|->|→|¬|∧|∨)", r" \1 ", expr)
    expr = re.sub(r"\s+", " ", expr)
    tokens = expr.split()
    normalized = []
    for tok in tokens:
        if tok.lower() in {"a", "e", "af", "ag", "ef", "eg", "u", "f", "g"}:
            normalized.append(tok.upper())
        else:
            normalized.append(tok)
    return " ".join(normalized)

# Split a list into train/val/test based on ratios
def split_dataset(data: Sequence[Tuple[str, str, Dict[str, str]]], ratios: Tuple[float, float, float] = (0.8, 0.1, 0.1), seed: int | None = None):
    if not math.isclose(sum(ratios), 1.0, abs_tol=1e-6):
        raise ValueError(f"Split ratios must sum to 1. Got {ratios}")
    n = len(data)
    if n == 0:
        return [], [], []
    if seed is not None:
        random.seed(seed)
    indices = list(range(n))
    random.shuffle(indices)
    train_end = int(ratios[0] * n)
    val_end = train_end + int(ratios[1] * n)
    train = [data[i] for i in indices[:train_end]]
    val = [data[i] for i in indices[train_end:val_end]]
    test = [data[i] for i in indices[val_end:]]
    return train, val, test


# ### 2.1 Load and preprocess your dataset
#
# Provide the path to your dataset in the variable `DATA_FILE` below. The cell loads the data, lifts atomic propositions, normalises formulas, splits the dataset into train/validation/test and saves these splits to JSONL files (`train.jsonl`, `val.jsonl`, `test.jsonl`).
#

# In[6]:


# Path to your dataset file (CSV/TSV or JSONL)
DATA_FILE = "/content/drive/MyDrive/Ghalya_CTL_Dataset-Joud-rim+ghalya.csv"  # Corrected path

# Load raw NL/TCTL pairs
raw_pairs = load_dataset(DATA_FILE)
print(f"Loaded {len(raw_pairs)} raw pairs.")

# Lift APs and normalise
processed = []
for nl, tctl in raw_pairs:
    lifted, mapping = lift_ap(tctl)
    normalised = normalize_formula(lifted)
    processed.append((nl, normalised, mapping))

# Preview a few examples
for idx in range(min(3, len(processed))):
    nl, tctl_proc, mp = processed[idx]
    print(f"Example {idx+1}:\n  NL: {nl}\n  Processed TCTL: {tctl_proc}\n  Mapping: {mp}\n")

# Split dataset
train_ratio, val_ratio, test_ratio = 0.8, 0.1, 0.1
train_samples, val_samples, test_samples = split_dataset(processed, ratios=(train_ratio, val_ratio, test_ratio), seed=42)

print(f"Train: {len(train_samples)}, Val: {len(val_samples)}, Test: {len(test_samples)}")

# # Save splits to JSONL
# import json

# def save_jsonl(filename, samples):
#     with open(filename, 'w', encoding='utf-8') as f:
#         for nl, tctl_proc, mp in samples:
#             json.dump({'nl': nl, 'tctl': tctl_proc, 'mapping': mp}, f)
#             f.write('\n')

# save_jsonl('train.jsonl', train_samples)
# save_jsonl('val.jsonl', val_samples)
# save_jsonl('test.jsonl', test_samples)
# print("Saved train.jsonl, val.jsonl, test.jsonl")
```

```python
import re
from dataclasses import dataclass
from typing import List, Tuple
import pandas as pd
from tqdm import tqdm
import pyuppaal as pyu

# ============================================================
# 1) AST DEFINITIONS FOR A SMALL TCTL FRAGMENT
# ============================================================

class Formula:
    """Base class for all formula nodes."""
    pass

@dataclass(frozen=True)
class Atom(Formula):
    name: str

@dataclass(frozen=True)
class Not(Formula):
    child: Formula

@dataclass(frozen=True)
class And(Formula):
    children: Tuple[Formula, ...]

@dataclass(frozen=True)
class Or(Formula):
    children: Tuple[Formula, ...]

@dataclass(frozen=True)
class AF(Formula):
    child: Formula

@dataclass(frozen=True)
class AG(Formula):
    child: Formula

@dataclass(frozen=True)
class EF(Formula):
    child: Formula

@dataclass(frozen=True)
class EG(Formula):
    child: Formula


# ============================================================
# 2) TOKENIZATION (WITH "ATOMS" THAT CAN CONTAIN CLOCK CONSTRAINTS)
# ============================================================

TEMPORAL_OPS = {"AF", "AG", "EF", "EG"}
BIN_OPS      = {"&&", "||", "->"}
UNARY_OPS    = {"!"}
PARENS       = {"(", ")"}
ALL_OPS      = TEMPORAL_OPS | BIN_OPS | UNARY_OPS | PARENS

def preprocess_for_parsing(formula: str) -> str:
    """
    Insert spaces around parens and basic operators so we can split safely,
    but we will later re-merge non-operator tokens into a single ATOM.
    """
    s = formula.strip()
    # multi-char operators first
    s = re.sub(r"(<=|>=|->)", r" \1 ", s)
    # single-char operators / parens
    s = re.sub(r"([()!<>=])", r" \1 ", s)
    # logical and/or shortcuts (if present)
    s = re.sub(r"(\&\&|\|\|)", r" \1 ", s)
    s = re.sub(r"\s+", " ", s)
    return s

def tokenize(formula: str) -> List[str]:
    """
    Convert a string formula into a list of tokens where:
      - UPPAAL/TCTL symbols (AF, AG, EF, EG, &&, ||, !, ->, (, )) are separate tokens
      - Everything else between them is merged into 'atom' tokens (e.g. 'x <= 5')
    """
    pre = preprocess_for_parsing(formula)
    raw_tokens = pre.split()

    merged_tokens: List[str] = []
    current_atom_parts: List[str] = []

    def flush_atom():
        nonlocal current_atom_parts, merged_tokens
        if current_atom_parts:
            merged_tokens.append(" ".join(current_atom_parts))
            current_atom_parts = []

    for tok in raw_tokens:
        t_upper = tok.upper()
        if t_upper in TEMPORAL_OPS or tok in BIN_OPS or tok in UNARY_OPS or tok in PARENS:
            # this is an operator or parenthesis
            flush_atom()
            # normalize temporal ops to uppercase
            if t_upper in TEMPORAL_OPS:
                merged_tokens.append(t_upper)
            else:
                merged_tokens.append(tok)
        else:
            # part of an atom (e.g. "x", "<=", "5")
            current_atom_parts.append(tok)

    flush_atom()
    return merged_tokens


# ============================================================
# 3) RECURSIVE DESCENT PARSER WITH PRECEDENCE
#    precedence:  lowest  ->  '->'
#                 then       '||'
#                 then       '&&'
#                 highest    unary (!, AF, AG, EF, EG)
# ============================================================

class Parser:
    def __init__(self, tokens: List[str]):
        self.tokens = tokens
        self.pos = 0

    def _peek(self):
        return self.tokens[self.pos] if self.pos < len(self.tokens) else None

    def _consume(self, expected=None):
        tok = self._peek()
        if tok is None:
            raise ValueError("Unexpected end of tokens")
        if expected is not None and tok != expected:
            raise ValueError(f"Expected '{expected}', got '{tok}'")
        self.pos += 1
        return tok

    def parse(self) -> Formula:
        if not self.tokens:
            raise ValueError("Empty formula")
        result = self._parse_implication()
        if self._peek() is not None:
            raise ValueError(f"Unconsumed tokens remain: {self.tokens[self.pos:]}")
        return result

    # lowest precedence
    def _parse_implication(self) -> Formula:
        left = self._parse_or()
        tok = self._peek()
        if tok == "->":
            self._consume("->")
            right = self._parse_implication()
            # a -> b  ===  !a OR b
            return Or((Not(left), right))
        return left

    def _parse_or(self) -> Formula:
        node = self._parse_and()
        while self._peek() == "||":
            self._consume("||")
            rhs = self._parse_and()
            node = Or((node, rhs))
        return node

    def _parse_and(self) -> Formula:
        node = self._parse_unary()
        while self._peek() == "&&":
            self._consume("&&")
            rhs = self._parse_unary()
            node = And((node, rhs))
        return node

    def _parse_unary(self) -> Formula:
        tok = self._peek()
        if tok is None:
            raise ValueError("Unexpected end in unary")

        # negation
        if tok == "!":
            self._consume("!")
            return Not(self._parse_unary())

        # temporal unary operators
        if tok in TEMPORAL_OPS:
            self._consume(tok)
            child = self._parse_unary()
            if tok == "AF":
                return AF(child)
            if tok == "AG":
                return AG(child)
            if tok == "EF":
                return EF(child)
            if tok == "EG":
                return EG(child)

        # parentheses
        if tok == "(":
            self._consume("(")
            inner = self._parse_implication()
            self._consume(")")
            return inner

        # otherwise it must be an ATOM
        self._consume()
        return Atom(tok)

def parse_formula_str(formula: str) -> Formula:
    tokens = tokenize(formula)
    parser = Parser(tokens)
    return parser.parse()


# ============================================================
# 4) CANONICALIZATION (NORMAL FORM) + EQUIVALENCE CHECK
# ============================================================

def canonicalize(f: Formula) -> Formula:
    """
    Normalize formula:
      - recursively canonicalize children
      - flatten And/Or
      - sort And/Or children by repr to ignore order
      - eliminate double negations
    This is not a full logical normalization, but it captures
    many common syntactic variations (commutativity/associativity).
    """
    if isinstance(f, Atom):
        return f

    if isinstance(f, Not):
        child = canonicalize(f.child)
        # double negation
        if isinstance(child, Not):
            return canonicalize(child.child)
        return Not(child)

    if isinstance(f, And):
        flat: List[Formula] = []
        for c in f.children:
            nc = canonicalize(c)
            if isinstance(nc, And):
                flat.extend(nc.children)
            else:
                flat.append(nc)
        # sort by repr for canonical order
        flat_sorted = tuple(sorted(flat, key=repr))
        if len(flat_sorted) == 1:
            return flat_sorted[0]
        return And(flat_sorted)

    if isinstance(f, Or):
        flat: List[Formula] = []
        for c in f.children:
            nc = canonicalize(c)
            if isinstance(nc, Or):
                flat.extend(nc.children)
            else:
                flat.append(nc)
        flat_sorted = tuple(sorted(flat, key=repr))
        if len(flat_sorted) == 1:
            return flat_sorted[0]
        return Or(flat_sorted)

    if isinstance(f, AF):
        return AF(canonicalize(f.child))
    if isinstance(f, AG):
        return AG(canonicalize(f.child))
    if isinstance(f, EF):
        return EF(canonicalize(f.child))
    if isinstance(f, EG):
        return EG(canonicalize(f.child))

    # fallback
    return f

def formulas_equivalent_str(f1: str, f2: str) -> bool:
    """
    Parse both strings into ASTs, canonicalize, and compare.
    Returns False if parsing fails for either.
    """
    try:
        ast1 = canonicalize(parse_formula_str(f1))
        ast2 = canonicalize(parse_formula_str(f2))
        return ast1 == ast2
    except Exception:
        return False


# ============================================================
# 5) CONVERSION FROM AST TO UPPAAL QUERY STRING
# ============================================================

def ast_to_uppaal(f: Formula, top_level: bool = True) -> str:
    """
    Convert our AST into a UPPAAL-like query string.
    Note: UPPAAL does not allow nested path modalities, so in
    complex cases this is an approximation of syntax, but for
    simple AF/AG/EF/EG usage it's fine.
    """
    # TOP-LEVEL PATH OPERATORS
    if top_level:
        if isinstance(f, AG):
            return "A[] " + ast_to_uppaal(f.child, top_level=False)
        if isinstance(f, AF):
            return "A<> " + ast_to_uppaal(f.child, top_level=False)
        if isinstance(f, EG):
            return "E[] " + ast_to_uppaal(f.child, top_level=False)
        if isinstance(f, EF):
            return "E<> " + ast_to_uppaal(f.child, top_level=False)
        # if there is no explicit temporal op, treat it as a state formula under A[]
        return "A[] " + ast_to_uppaal(f, top_level=False)

    # STATE-FORMULA LEVEL
    if isinstance(f, Atom):
        return f.name

    if isinstance(f, Not):
        return f"!({ast_to_uppaal(f.child, top_level=False)})"

    if isinstance(f, And):
        return "(" + " && ".join(ast_to_uppaal(c, top_level=False) for c in f.children) + ")"

    if isinstance(f, Or):
        return "(" + " || ".join(ast_to_uppaal(c, top_level=False) for c in f.children) + ")"

    if isinstance(f, AF):
        return "A<> " + ast_to_uppaal(f.child, top_level=False)
    if isinstance(f, AG):
        return "A[] " + ast_to_uppaal(f.child, top_level=False)
    if isinstance(f, EF):
        return "E<> " + ast_to_uppaal(f.child, top_level=False)
    if isinstance(f, EG):
        return "E[] " + ast_to_uppaal(f.child, top_level=False)

    # fallback
    return str(f)


# ============================================================
# 6) OPTIONAL: USE UPPAAL TO CHECK SATISFACTION OF A QUERY
# ============================================================

def verify_formula(model: pyu.UModel, uppaal_query: str) -> bool:
    """
    Return True if `uppaal_query` holds on `model`, False otherwise.
    Expects `uppaal_query` to already be in UPPAAL syntax.
    """
    try:
        m_copy = model.copy()
        m_copy.queries = [uppaal_query]
        res = m_copy.verify()
        return ("Formula is satisfied" in res) or ("Property is satisfied" in res)
    except Exception:
        return False  # invalid syntax or unsupported operator


# ============================================================
# 7) EVALUATION LOOP USING SEMANTIC (AST-BASED) MATCH
# ============================================================

results = []
for idx, (_, row) in enumerate(tqdm(test_df.iterrows(), total=len(test_df), desc="Evaluating")):
    nl_input   = row['refined NL']
    expected   = row['TCTL']
    predicted  = generate_prediction(nl_input, task_prefix)

    # ---- SEMANTIC (AST-BASED) EQUIVALENCE ----
    sem_match = formulas_equivalent_str(expected, predicted)

    # ---- Optional: convert to UPPAAL format + check satisfiability on a model ----
    try:
        exp_ast  = canonicalize(parse_formula_str(expected))
        pred_ast = canonicalize(parse_formula_str(predicted))

        exp_up   = ast_to_uppaal(exp_ast)   # UPPAAL query for expected
        pred_up  = ast_to_uppaal(pred_ast)  # UPPAAL query for predicted

        exp_ok   = verify_formula(model, exp_up)
        pred_ok  = verify_formula(model, pred_up)
    except Exception:
        exp_up = "<parse error>"
        pred_up = "<parse error>"
        exp_ok = False
        pred_ok = False

    print(f"\nCase {idx+1}:")
    print(f"  Input NL               : {nl_input}")
    print(f"  Expected TCTL          : {expected}")
    print(f"  Predicted TCTL         : {predicted}")
    print(f"  Canonical semantic match?: {sem_match}")
    print(f"  Expected (UPPAAL query): {exp_up}")
    print(f"  Predicted (UPPAAL query): {pred_up}")
    print(f"  Expected satisfied on model?: {exp_ok}")
    print(f"  Predicted satisfied on model?: {pred_ok}")

    results.append({
        'input': nl_input,
        'expected': expected,
        'predicted': predicted,
        'semantic_match': sem_match,
        'expected_uppaal': exp_up,
        'predicted_uppaal': pred_up,
        'expected_satisfied': exp_ok,
        'predicted_satisfied': pred_ok,
    })

results_df = pd.DataFrame(results)
semantic_accuracy = results_df['semantic_match'].mean()
print(f"\nAST-based semantic accuracy: {semantic_accuracy:.2%}")
print(f"Semantic matches: {results_df['semantic_match'].sum()}/{len(results_df)}")
```
